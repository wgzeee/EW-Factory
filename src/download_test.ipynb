```python
start_time = time.time()  # 记录开始时间
client.retrieve(dataset, request).download("era5/test1month.nc")
end_time = time.time()  # 记录结束时间

elapsed_time = end_time - start_time  # 计算运行时间
print(f"代码运行时间: {elapsed_time:.2f} 秒")
```
